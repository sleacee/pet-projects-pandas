Руководитель отдела сопровождения клиентов хочет получить информацию о том, насколько хорошо сотрудники его отдела (аккаунт-менеджеры) справляются с одной из своих основных задач – пролонгацией договоров с клиентами. От аналитика он хочет получить отчет о пролонгациях сотрудников за 2023 год.

В компании используется два коэффициента пролонгации:
Для проектов пролонгированных в первый месяц – отношении суммы отгрузки проектов пролонгированных в первый месяц после завершения к сумме отгрузки последнего месяца реализации всех завершившихся в прошлом месяце проектов.
Для проектов, пролонгированных во второй месяц – отношение суммы отгрузки проектов, пролонгированных во второй месяц к сумме отгрузки последнего месяца проектов, не пролонгированных в первый.

То есть, если нам нужно понять, насколько хорошо менеджер пролонгировал в мае, необходимо посчитать:
Сумму отгрузки проектов, завершившихся в апреле (за апрель) и сумму отгрузки тех проектов завершившихся в апреле, у которых есть отгрузка в мае (за май). Коэффициент – отношение второй суммы к первой.
Сумму проектов, завершившихся в марте, у которых нет отгрузки в апреле (за март) и сумму отгрузки тех проектов, завершившихся в марте, у которых нет отгрузки в апреле но есть в мае (за май). Коэффициент – отношение второй суммы к первой.

Необходимо:
Рассчитать коэффициенты пролонгации для каждого менеджера и для всего отдела в целом
за каждый месяц и
за год
Сформировать аналитический отчет в гугл-таблицах или excel, на основании которого руководитель отдела будет принимать управленческие решения (обязательно присутствие коэффициентов пролонгации, в остальном отчет можно дополнить на свое усмотрение – дополнительные метрики и визуализация приветствуются).


```python
#подключаем используемые библиотеки и скачиваем в проект используемые csv
import pandas as pd
import numpy as np
! gdown --id 1kSg3Ln2L7KyWk2ccniphlsqDuBZD4ytQ
! gdown --id 1lUfXK-WML9LlNjzmI9IncQWLsM_v21wx
```

```python
#загружаем даныне из csv
financial_data = pd.read_csv('/content/financial_data.csv')
prolongations = pd.read_csv('/content/prolongations.csv')
```

```python
#financial_data чистим от текста, nan'ов и приводим числа в нормальный вид
def clean_currency(value):
    if pd.isna(value) or str(value).strip().lower() in ['стоп', 'в ноль', '', 'end']:
        return 0.0
    clean_val = str(value).replace('\xa0', '').replace(',', '.')
    return float(clean_val)


months = ['Ноябрь 2022', 'Декабрь 2022', 'Январь 2023',
       'Февраль 2023', 'Март 2023', 'Апрель 2023', 'Май 2023', 'Июнь 2023',
       'Июль 2023', 'Август 2023', 'Сентябрь 2023', 'Октябрь 2023',
       'Ноябрь 2023', 'Декабрь 2023', 'Январь 2024', 'Февраль 2024',
       ]
for col in months:
    financial_data[col] = financial_data[col].apply(clean_currency)
```

```python
#поскольку причины дублей айди нас не интересуют - группируем данные из financial_data по айди и суммируем отгрузки
fin_grouped = financial_data.groupby('id')[months].sum().reset_index()
fin_grouped
```

```python
#совмещаем данные из полученной таблицы с данными по пролонгациям
df = prolongations.merge(fin_grouped, on='id', how='left')
df
```

```python
#собираем список менеджеров
managers = df['AM'].unique()
managers
```

```python
#сюда будем записывать рассчеты для каждого менеджера
result = []
#отчет за 2023 год - берем только месяца 23 года из уже созданного массива с месяцами
months_target = months[2:14]
#итератор для отслеживания 2 предыдущих месяцев
i = 0

#пройдемся по всем месяцам из таргета
for col in months_target:
  #предыдущий месяц
  prev_1 = months[i+1]
  #позапрошлый месяц
  prev_2 = months[i]

  #датафрейм с данными, где месяц пролонгации совпадает с предыдущим месяцем для рассчета первого коэффициента пролонгации
  proj_m1 = df[df['month'].str.lower() == prev_1.lower()]

  #для второго коэффициента пролонгации - сравнение месяца с позапрошлым месяцем и отсутствие отгрузки в прошлом месяце
  proj_m2 = df[(df['month'].str.lower() == prev_2.lower()) & (df[prev_1] == 0)]

  #далее в исследуемом месяце проходимся по всем менеджерам
  for am in managers:

    #фильтруем датафрейм по выбранному аккаунт-менеджеру
    am_m1 = proj_m1[proj_m1['AM'] == am]
    #в знаменатель считаем сумму отгрузок по предыдущему месяцу
    m1_sum_1 = am_m1[prev_1].sum()
    #в числитель считаем сумму отгрузок по предыдущему месяцу с учетом наличия отгрузки в текущем месяце
    m1_sum_2 = am_m1[am_m1[col] > 0][prev_1].sum()
    #из полученных значений считаем коэффициент
    cr1 = m1_sum_2 / m1_sum_1

    #аналогично для второго коэффициента
    am_m2 = proj_m2[proj_m2['AM'] == am]
    m2_sum_1 = am_m2[prev_2].sum()
    m2_sum_2 = am_m2[am_m2[col] > 0][prev_2].sum()
    cr2 = m2_sum_2 / m2_sum_1

    #добавляем посчитанные значения в список
    result.append({
      'Month' : col,
      'AM' : am,
      'CR1' : cr1,
      'CR2' : cr2,
      'To_m1' : m1_sum_1,
      'End_m1' : m1_sum_2,
      'To_m2' : m2_sum_1,
      'End_m2' : m2_sum_2,
    })
  #увеличеваем итератор на один для перехода на следующий месяц
  i+=1





```

```python
#датафрейм с результатами
res = pd.DataFrame(result)
res = res.fillna(0)

#датафрейм с значениями коэффициента за год для каждого менеджера
manager_2023 = res.groupby('AM', sort=False).agg({'To_m1': 'sum', 'End_m1': 'sum','To_m2': 'sum','End_m2': 'sum',})
manager_2023['Y_CR1'] = manager_2023['End_m1'] / manager_2023['To_m1']
manager_2023['Y_CR2'] = manager_2023['End_m2'] / manager_2023['To_m2']

#коэффициенты за год для всего отдела
total_2023_cr1 = manager_2023['End_m1'].sum() / manager_2023['To_m1'].sum()
total_2023_cr2 = manager_2023['End_m2'].sum() / manager_2023['To_m2'].sum()

#коэффициенты для всего отдела по месяцам
total_month = res.groupby('Month', sort=False).agg({'To_m1': 'sum', 'End_m1': 'sum','To_m2': 'sum','End_m2': 'sum',})
total_month['CR1'] = total_month['End_m1'] / total_month['To_m1']
total_month['CR2'] = total_month['End_m2'] / total_month['To_m2']
res
```

```python
#выгружаем данные в эксель для формирования отчета
res.to_excel('Менеджеры по месяцам.xlsx', index=False)
```
